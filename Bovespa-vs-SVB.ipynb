```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Baixando os preços das ações do Silicon Valley Bank do último ano
df = yf.download('SIVB', start='2022-01-01', end='2023-02-28')

# Exibindo os dados baixados - estamos interessados nas colunas Close (Preço de Fechamento) e Adj Close (Preço de Fechamento Ajustado)
print(df)

# Plotando o gráfico Tempo x PFA
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Adj Close'])
plt.title('Preços de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento Ajustado')
plt.show()

# Plotando o gráfico Tempo x PF
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Close'])
plt.title('Preços de Fechamento do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.show()

# Comparando PF e PFA com o tempo
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Close'], label='Preço de Fechamento')
plt.plot(df.index, df['Adj Close'], label='Preço de Fechamento Ajustado')
plt.title('Preços de Fechamento e Preços de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.legend()
plt.show()

# Plotando o gráfico Tempo x Altas das Ações
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['High'])
plt.title('Altas das ações do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Altas das ações')
plt.show()

# Comparando Altas das Ações com Preços de Fechamento Ajustados
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['High'], label='Altas das ações')
plt.plot(df.index, df['Adj Close'], label='Preço de Fechamento Ajustado')
plt.title('Altas das Ações e Preços de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.legend()
plt.show()

# Plotando o gráfico Tempo x Baixas das Ações
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'])
plt.title('Baixas das ações do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Baixas das ações')
plt.show()

# Comparando Baixas das Ações com Preços de Fechamento Ajustados
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'], label='Baixas das ações')
plt.plot(df.index, df['Adj Close'], label='Preço de Fechamento Ajustado')
plt.title('Baixas das Ações e Preços de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.legend()
plt.show()

# Comparando Baixas e Altas das Ações 
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'], label='Baixas das ações')
plt.plot(df.index, df['High'], label='Altas das ações')
plt.title('Baixas e Altas das ações do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.legend()
plt.show()

# Comparando Baixas e Altas das Ações com os Preços de Fechamento Ajustados 
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'], label='Baixas das ações')
plt.plot(df.index, df['High'], label='Altas das ações')
plt.plot(df.index, df['Adj Close'], label='Preço de Fechamento Ajustado')
plt.title('Baixas, Altas e Preços de Fechamento Ajustados das ações do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.legend()
plt.show()
```
