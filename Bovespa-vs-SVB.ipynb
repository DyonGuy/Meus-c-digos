```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Baixando os preos das aes do Silicon Valley Bank do ltimo ano
df = yf.download('SIVB', start='2022-01-01', end='2023-02-28')

# Exibindo os dados baixados - estamos interessados nas colunas Close (Preo de Fechamento) e Adj Close (Preo de Fechamento Ajustado)
print(df)

# Plotando o grfico Tempo x PFA
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Adj Close'])
plt.title('Preos de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preo de Fechamento Ajustado')
plt.show()

# Plotando o grfico Tempo x PF
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Close'])
plt.title('Preos de Fechamento do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preo de Fechamento')
plt.show()

# Comparando PF e PFA com o tempo
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Close'], label='Preo de Fechamento')
plt.plot(df.index, df['Adj Close'], label='Preo de Fechamento Ajustado')
plt.title('Preos de Fechamento e Preos de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preo')
plt.legend()
plt.show()

# Plotando o grfico Tempo x Altas das Aes
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['High'])
plt.title('Altas das aes do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Altas das aes')
plt.show()

# Comparando Altas das Aes com Preos de Fechamento Ajustados
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['High'], label='Altas das aes')
plt.plot(df.index, df['Adj Close'], label='Preo de Fechamento Ajustado')
plt.title('Altas das Aes e Preos de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preo')
plt.legend()
plt.show()

# Plotando o grfico Tempo x Baixas das Aes
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'])
plt.title('Baixas das aes do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Baixas das aes')
plt.show()

# Comparando Baixas das Aes com Preos de Fechamento Ajustados
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'], label='Baixas das aes')
plt.plot(df.index, df['Adj Close'], label='Preo de Fechamento Ajustado')
plt.title('Baixas das Aes e Preos de Fechamento Ajustados do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preo')
plt.legend()
plt.show()

# Comparando Baixas e Altas das Aes 
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'], label='Baixas das aes')
plt.plot(df.index, df['High'], label='Altas das aes')
plt.title('Baixas e Altas das aes do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preo')
plt.legend()
plt.show()

# Comparando Baixas e Altas das Aes com os Preos de Fechamento Ajustados 
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['Low'], label='Baixas das aes')
plt.plot(df.index, df['High'], label='Altas das aes')
plt.plot(df.index, df['Adj Close'], label='Preo de Fechamento Ajustado')
plt.title('Baixas, Altas e Preos de Fechamento Ajustados das aes do Silicon Valley Bank (2022-2023)')
plt.xlabel('Data')
plt.ylabel('Preo')
plt.legend()
plt.show()
```
