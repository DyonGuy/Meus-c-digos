```python
# Vamos explicitar, inicialmente, o significado de cada comando

# Comando "and" tem o valor lógico de "E" 

# Comando "or" tem o valor lógico de "Ou"

# Comando "not" significa a negação de uma sentença, alterando o valor de "true" para "false" e vice-versa (~P). 

# Por simplicidade, podemos definir os resultados das operações lógicas da seguinte maneira:

p = True

q= False

#Abaixo montaremos as tabelas


#############################################################################################################
#################################### TABELA AND #############################################################


print("p\tq\tp and q") # Este será o cabeçalho da nossa tabela-verdade. O comando \t é o espaço entre as célula
print("-\t-\t----------") # Aqui será a separação tracejadada entre o cabeçalo e as célula da tabela

print("{}\t{}\t{}".format(p, not q, p and not q))
print("{}\t{}\t{}".format(p,  q, p and q))
print("{}\t{}\t{}".format(not p, not q, not p and not q))
print("{}\t{}\t{}".format(not p, q, not p and q))

print("-------\t---------\t----------")



############################################################################################################
################################### TABELA OR ##############################################################


print("p\tq\tp or q") # Este será o cabeçalho da nossa tabela-verdade. O comando \t é o espaço entre as célula
print("-\t-\t----------") # Aqui será a separação tracejadada entre o cabeçalo e as célula da tabela

print("{}\t{}\t{}".format(p, not q, p or not q))
print("{}\t{}\t{}".format(p, q, p or not q))
print("{}\t{}\t{}".format(not p, not q, not p or not q))
print("{}\t{}\t{}".format(not p, q, not p or q))

print("-------\t---------\t----------")




############################################################################################################
################################### TABELA SE, ... ENTÃO.... ###############################################

# Esta será um diferente das demais, pois o comando "if" não pode simplesmente ser introduzido da mesma forma que os comandos "and" e "or"

combinations = [(p, not q), (p, q), (not p, not q), (not p, q)] #Estão são as possíveis combinações que entrarão nas células

print("p | q | p -> q")
print("--|---|-------")

# Escrevendo "se p, então q":

for p, q in combinations:
    resultado = q if p else True
    print(f"{p} | {q} | {resultado}")
print("-------\t---------\t----------")
############################################################################################################
################################### TABELA ...SE, E SOMENTE SE.... #########################################


combinations = [(p, not q), (p, q), (not p, not q), (not p, q)] #Estão são as possíveis combinações que entrarão nas células

print("p | q | p <-> q")
print("--|---|-------")

# Escrevendo "se p, então q":

for p, q in combinations:
    resultado = (q if p else True) and (p if q else True)
    print(f"{p} | {q} | {resultado}")


print("-------\t---------\t----------")


############################################################################################################
################################### TABELA OR... OR... #########################################


combinations = [(p, not q), (p, q), (not p, not q), (not p, q)] #Estão são as possíveis combinações que entrarão nas células

print("p | q | or p or q")
print("--|---|-------")

# Escrevendo "ou p, ou q":

for p, q in combinations:
    resultado = not((q if p else True) and (p if q else True))
    print(f"{p} | {q} | {resultado}")


print("-------\t---------\t----------")
```
